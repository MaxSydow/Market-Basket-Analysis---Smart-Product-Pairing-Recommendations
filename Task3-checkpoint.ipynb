# Market Basket Analysis

Proportions of single products can be calculated by summing up all instances of each then dividing by the total.  This would indicate a probability of an item being purchased.  The probability of 1 or more items being purchased is also known as a metric called Support.  Furthermore, the conditional probability of purchasing one product given that another had been purchased can also be calculated.  Another name for this metric is Confidence.  A third useful number is called Lift, which indicates the degree of randomness in Confidence.  (Datacamp)

Support(X) = freq(X) / N

Confidence(X, Y) = Support(X, Y)  / Support(X)

Lift(X, Y) = Support(X, Y)  / Support(X)Support(Y)

The essential idea behind Market Basket Analysis (MBA) is to use these 3 metrics to form Association Rules.  These rules take an if then form of Antecedent => Consequent.  In other words the 3 metrics above allow for the likelihood of a consequent product being purchased given that an antecedent was purchased.  These rules can only be obtained from existing data, so they don't necessarily indicate causality.  Nevertheless, history can be valuable in attempting to predict future patterns and the purchase history in this data set covers 2 years.

There are 20 unique items in this data set, and the number of combinations of products purchased increases astronomically as more items are considered. Thankfully, there is an implimentable algorithm that can be used to narrow down the number of combinations.  According to Datacamp "The Apriori algorithm is structured around the idea that we should retain items that are frequent -- that is, exceed some minimal level of support. The Apriori principle states that subsets of frequent sets must also be frequent. The Apriori algorithm uses this principle to retain frequent sets and prune those that cannot be said to be frequent."

In general confidence, which ranges from 0 to 1, should be high.  Also, a lift greater than 1 implies that 2 items in a transaction exceeds their random occurance together.  Python also has an *association_rules* function that ranks association rules according to support, confidence and lift.  

# Data Preparation and Analysis

In order to calculate the 3 metrics and use them as parameters in the Apriori algorithm and association_rules function some data cleaning and pre-processing is required.  The raw data, read in as a dataframe, contains blank rows in between each row of customer transactions.  The columns are labeled Item01 - Item20, but do not indicate the same actual item.  Entries are filled in successive columns from 1st item purchased by a customer to last.  Most rows have blank entries after a certain number of items.  The blank rows were dropped, while the remaining blank values were converted to 0s.  

The cleaned dataframe was then converted to a list of lists of items by transaction.  A single list item summarizes a single customers history of transactions.  This list was then one-hot encoded so that each item can be counted and the metric calculations computed.

```python
# one-hot encode the list

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori

# one-hot encode
encoder = TransactionEncoder()
onehot = encoder.fit(tran_list).transform(tran_list)
onehot = pd.DataFrame(onehot, columns = encoder.columns_)
```

Minimal support can be used in the Apriori algorithm and association_rules method to obtain a relatively large sample of rules.  The 3 metrics are included in association_rule's output.  A scatterplot of confidence and support can then be made.  Each point can then be weighted by size according to lift.  

Datacamp sites work by Bayardo and Agrawal that proved the best performing association rules have metrics that lie along the edges of such a scatterplot.  Finding the largest point by lift along the edges can help narrow down the number of association rules returned. 

```python
# Import seaborn under its standard alias
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import association_rules

# Apply the Apriori algorithm with a support value of 0.0075
frequent_itemsets = apriori(onehot, min_support = 0.0075, 
                         use_colnames = True, max_len = 2)

# Generate association rules without performing additional pruning
rules = association_rules(frequent_itemsets, metric = "support", 
                          min_threshold = 0.0)

# Generate scatterplot using support and confidence
sns.scatterplot(x = "support", y = "confidence", 
                size = "lift", data = rules)
plt.show()
```

It looks like the optimal metric values would occur on the left hand edge of the plot in the vicinity where confidence = 0.4.  After some trial and error, the following minimum values produced a list of 4 rules.  

Support: 0.008

Confidence: 0.4

Lift: 1.5

```python
# Recover association rules using a minium support threshold of 0.008
rules = association_rules(frequent_itemsets, metric = 'support', min_threshold = 0.008)

# Apply a 0.002 antecedent support threshold, 0.60 confidence threshold, and 2.50 lift threshold
filtered_rules = rules[(rules['support'] > 0.008) &
                       (rules['confidence'] > 0.4) &
                       (rules['lift'] > 1.5)]

# Print remaining rule
print(filtered_rules)
```

# Summary and Implications

A table of the top 3 association rules, along with their support, confidence and lift values can be extracted from the filtered_rules dataframe.

```python
filtered_rules[['antecedents', 'consequents', 'antecedent support', 'consequent support', 'support', 'confidence', 'lift']].head(3)
```

These rules can be summarized as:

1. 10ft iPHone Charger Cable 2 Pack => Dust-Off Compressed Gas 2 pack

2. 3A USB Type C Cable 3 pack 6FT => Dust-Off Compressed Gas 2 pack

3. FEIYOLD Blue light Blocking Glasses => Dust-Off Compressed Gas 2 pack

Each antecedent's support is not very high; the consequent's support is much higher.  Customers have purchased a higher proportion of compressed gas than the cables or glasses.  The support for both items in each rule is close to the support for purchasing the gas, so customers are about as likely to purchase both by random chance.  Confidence for each rule is much higher than support for both items, meaning that customers are much more likely to purchase the consequent item, given that they have purchased one of the atecedent items.  Furthermore the lift values for each rule is greater than 1, so these rules are good ones in that their occurance is not due to random chance.  

In practicality, what these rules say is that if customers buy one of these types of cables or this particular glasses item then they are many times more likely to also purchase a 2 pack of compressed gas.  A physical store may find an increase in the 2 pack gas item sales if these items were placed near a cable or glasses/peripheral section.  An online retailer could suggest the compressed gas when customers add one or more of the three antecedent items to their cart.  The results of market basket analysis suggest a higher likelihood that a customer would then add the compressed gas item to their cart as well.
